Methods to open file in python.

```python
def dataset(filename):
    file_obj = open(filename,'r')
    string_file = file_obj.read()
    list_data = string_file.split()
    return list_data

data = dataset('guns.csv')
data[1:5]
```

```python
import csv
def dataset(filename):
    file_obj = open(filename,'r')
    csvreader = csv.reader(file_obj)
    data = list(csvreader)
    return data
data = dataset('guns.csv')
data[1:5]
```

Exploring Gun Deaths By Year.

```python
years = [row[1] for row in data]
years
```

```python
year_counts = {}
for year in years:
    if year in year_counts:
        year_counts[year] = year_counts[year] + 1
    else:
        year_counts[year] = 1
print(year_counts)
        
      
```

Exploring Gun Deaths By Month And Year.

```python
import datetime
dates = [datetime.datetime(year=int(row[1]), month=int(row[2]), day=1) for row in data]
dates[:5]
```

```python
date_counts = {}
for row in dates:
    if row in date_counts:
        date_counts[row] = date_counts[row]+1
    else:
        date_counts[row] = 1
print(date_counts)
```

Exploring Gun Deaths By Race And Sex.

```python
sex_data = [row[5] for row in data]
race_data = [row[7] for row in data]
```

```python
sex_counts = {}
for sex in sex_data:
    if sex in sex_counts:
        sex_counts[sex]+=1
    else:
        sex_counts[sex]=1
```

```python
sex_counts
```

```python
race_counts = {}
for race in race_data:
    if race not in race_counts:
        race_counts[race] = 0
    race_counts[race] += 1
race_counts
```
